```python
import matplotlib.pyplot as plt
import numpy as np

import ehrapy as ep
```

```python
%load_ext autoreload
%autoreload 2
```

```python
current_notebook_dir = %pwd

_TEST_IMAGE_PATH = f"{current_notebook_dir}/../plot/_images"
```

```python
adata_full = ep.dt.mimic_2(encoded=False)
adata = adata_full[:200].copy()
```

```python
co2_lm_result = ep.tl.ols(
    adata, var_names=["pco2_first", "tco2_first"], formula="tco2_first ~ pco2_first", missing="drop"
).fit()
ep.pl.ols(
    adata,
    x="pco2_first",
    y="tco2_first",
    ols_results=[co2_lm_result],
    ols_color=["red"],
    xlabel="PCO2",
    ylabel="TCO2",
    show=False,
)

plt.gcf().set_size_inches(8, 6)
plt.gcf().subplots_adjust(left=0.2, right=0.8, bottom=0.2, top=0.8)

plt.gcf().savefig(f"{_TEST_IMAGE_PATH}/ols_expected.png", dpi=80)
```
