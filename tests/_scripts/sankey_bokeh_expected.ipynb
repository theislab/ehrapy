```python
layer = np.array(
    [
        [[0, 1, 2, 1, 2], [1, 2, 1, 2, 0]],
        [[1, 2, 0, 2, 1], [2, 1, 2, 1, 2]],
        [[2, 0, 1, 2, 0], [2, 0, 1, 1, 2]],
        [[1, 2, 1, 0, 1], [0, 2, 1, 2, 0]],
        [[0, 2, 1, 2, 2], [1, 2, 1, 0, 2]],
    ]
)


edata_time = ed.EHRData(
    layers={"layer_1": layer},
    obs=pd.DataFrame(index=["patient_1", "patient_2", "patient_3", "patient_4", "patient_5"]),
    var=pd.DataFrame(index=["treatment", "disease_flare"]),
    tem=pd.DataFrame(index=["visit_0", "visit_1", "visit_2", "visit_3", "visit_4"]),
)
```

```python
ep.pl.plot_sankey_time(
    edata_time,
    columns=["disease_flare"],
    layer="layer_1",
    state_labels={0: "no flare", 1: "mid flare", 2: "severe flare"},
)
```
