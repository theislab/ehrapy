```python
import ehrdata as ed
import holoviews as hv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import ehrapy as ep
```

```python
current_notebook_dir = %pwd

_TEST_IMAGE_PATH = f"{current_notebook_dir}/../plot/_images"
```

```python
edata = ed.dt.diabetes_130_fairlearn(columns_obs_only=["gender", "race"])[:100]
```

```python
sankey = ep.pl.plot_sankey(edata, columns=["gender", "race"], backend="matplotlib", show=False)
fig = hv.render(sankey, backend="matplotlib")

fig.savefig(f"{_TEST_IMAGE_PATH}/sankey_expected.png", dpi=80)
```

```python
ep.pl.plot_sankey(edata, columns=["gender", "race"], backend="bokeh", show=False)
```

```python
layer = np.array(
    [
        [[1, 0, 1], [0, 1, 0]],  # patient 1: treatment, disease_flare
        [[0, 1, 1], [1, 0, 0]],  # patient 2: treatment, disease_flare
        [[1, 1, 0], [0, 0, 1]],  # patient 3: treatment, disease_flare
    ]
)

edata_time = ed.EHRData(
    layers={"layer_1": layer},
    obs=pd.DataFrame(index=["patient_1", "patient_2", "patient_3"]),
    var=pd.DataFrame(index=["treatment", "disease_flare"]),
    tem=pd.DataFrame(index=["visit_0", "visit_1", "visit_2"]),
)
```

```python
sankey_time = ep.pl.plot_sankey_time(
    edata_time,
    columns=["disease_flare"],
    layer="layer_1",
    state_labels={0: "no flare", 1: "flare"},
    backend="matplotlib",
)

fig = hv.render(sankey_time, backend="matplotlib")
fig.savefig(f"{_TEST_IMAGE_PATH}/sankey_time_expected.png", dpi=80)
```
