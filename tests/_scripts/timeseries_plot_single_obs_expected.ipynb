```python
import ehrdata as ed
import numpy as np
import pandas as pd
from ehrdata.core.constants import DEFAULT_TEM_LAYER_NAME

import ehrapy as ep
```

```python
%load_ext autoreload
%autoreload 2

current_notebook_dir = %pwd
_TEST_IMAGE_PATH = f"{current_notebook_dir}/../plot/_images"
```

```python
edata = ed.dt.ehrdata_blobs(
    n_variables=4,
    n_observations=10,
    base_timepoints=100,
    layer=DEFAULT_TEM_LAYER_NAME,
)

edata.var.index = ["feature1", "feature2", "feature3", "feature4"]
edata.obs["subject_id"] = [f"patient{i}" for i in range(1, edata.n_obs + 1)]
```

```python
edata.obs
```

```python
# timeseries of one observation with row index
ax1 = ep.pl.plot_timeseries(
    edata,
    obs_id=2,
    keys=["feature1", "feature2", "feature3"],
    layer=DEFAULT_TEM_LAYER_NAME,
    tem_time_key="timepoint",
    show=False,
)

ax1.figure.savefig(f"{_TEST_IMAGE_PATH}/plot_timeseries_one_obs_row_index_expected.png", dpi=80)
```

```python
# timeseries of one observation with obs id

ax2 = ep.pl.plot_timeseries(
    edata,
    obs_id="patient3",
    keys=["feature1", "feature2", "feature3"],
    obs_id_key="subject_id",
    layer=DEFAULT_TEM_LAYER_NAME,
    tem_time_key="timepoint",
    show=False,
)

ax2.figure.savefig(f"{_TEST_IMAGE_PATH}/plot_timeseries_one_obs_id_expected.png", dpi=80)
```
